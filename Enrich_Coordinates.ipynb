```python
import googlemaps
gmaps = googlemaps.Client(MAPS_KEY)
```

```python
def get_coordinates(plz_string):
    location = gmaps.geocode(plz_string)
    try:
        lon = location[0]["geometry"]["location"]["lng"]
        lat = location[0]["geometry"]["location"]["lat"]
    except:
        lon = None
        lat = None
        print("Something went wrong with: {}".format(plz_string))
    return lat, lon
```

```python
if 'location' in df.columns:
    pass
else:
    df['location'] = df["PLZ"].apply(lambda x: str(x) + ", Germany")
```

```python
df['lat / lon'] = df['location'].apply(lambda x: get_coordinates(x))
```
